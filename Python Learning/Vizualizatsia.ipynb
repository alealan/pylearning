```python
fig = plt.figure(figsize= (14,5))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(122)
ax3 = fig.add_subplot(223)

plt.plot(randn(100).cumsum(), 'k--')
f = ax1.hist(randn(100), bins=20, color="y", alpha=0.9)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
```

```python
fig, axec = plt.subplots(2, 3)
fig.set_figheight(15)
fig.set_figwidth(15)

axec[0,1].scatter(np.arange(30), np.arange(30) + 3 * randn(30))
axec[1,1].bar(np.arange(10), randn(10))
```

```python
axec
```

```python
fig, axec = plt.subplots(2, 2, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)
for i in range(2):
    for g in range(2):
        axec[i, g].hist(randn(500), bins=50, color='r', alpha=0.5)
```
